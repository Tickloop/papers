```python
import json
papers = []
with open('../data/acl.json', 'r', encoding='utf-8') as f:
    papers = json.load(f)
```

```python
venues = set()
for paper in papers:
    if 'publisher' in paper:
        venues.add(f"{paper['publisher']}-{paper['year']}")
    
    elif 'booktitle' in paper:
        venues.add(f"book-{paper['booktitle']}-{paper['year']}")
```

```python
for venue in sorted(venues):
    print(venue)
```

```python
acl_sub_books = set()
def get_paper_venue(paper: dict):
    venue = ''
    if publisher := paper.get('publisher'):
        venue = venue + publisher

    if book := paper.get('booktitle'):
        venue = venue + f"-{book}"

    if not venue:
        venue = paper.get('title')
    
    if year := paper.get('year'):
        venue = venue + f"-{year}"
    return venue

for paper in papers:
    if venue := get_paper_venue(paper):
        acl_sub_books.add(venue)

for book in sorted(acl_sub_books):
    print(book)
```

```python
with open('../data/acl-venues.json', 'r', encoding='utf-8') as f:
    venues = json.load(f)


matched_papers = []
for paper in papers:
    paper_venue = get_paper_venue(paper)

    matched_venue = {}
    for venue in venues:
        if venue['name'] in paper_venue:
            matched_venue = venue
        
    if matched_venue:
        matched_papers.append({
            **paper,
            'venue_name': matched_venue['name'],
            'venue_id': matched_venue['abbr'],
            'venue_url': matched_venue['url']
        })
```
