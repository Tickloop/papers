```python
from bs4 import BeautifulSoup

with open('./acl-venue-list.html', 'r') as f:
    html_doc = f.read()
soup = BeautifulSoup(html_doc, 'html.parser')
venues = [{
    'name': a.text.strip(),
    'url': f"https://aclanthology.org{a['href']}",
    'abbr': a.find_next('span', _class='badge-secondary').text.strip(),
    'volume_count': int(a.find_next('span', _class='badge-info').text.split()[0]),
} for a in soup.find_all('a', href=True) if a['href']]
```

```python
links = soup.find_all('a', href=True)
print(len(links))

abbrs = soup.find_all('span', class_='badge-secondary')
print(len(abbrs))

volumes = soup.find_all('span', class_='badge-info')
print(len(volumes))
```

```python
links[0]
```

```python
venues = []
for idx in range(len(links)):
    venues.append({
        'url': f'https://aclanthology.org{links[idx]["href"]}',
        'name': links[idx].text.strip(),
        'abbr': abbrs[idx].text.strip(),
        'volume_count': int(volumes[idx].text.strip().split()[0]),
    })
```

```python
import json
with open('../data/acl-venues.json', 'w', encoding='utf-8') as f:
    json.dump(venues, f)
```
